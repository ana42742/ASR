```python
# pipeline is the fastest way to create an ASR web app
from transformers import pipeline

# p = pipeline("automatic-speech-recognition", model_name)
```

```python
p = pipeline("automatic-speech-recognition")
```

```python
# build web app
```

```python
import gradio as gr
import time
```

```python
# def transcribe(audio):
#     """ Speech to text function using the pipeline that we defined """
#     text = p(audio)["text"]
#     return text
```

```python
# chunking audio into smaller pieces and passing it into transcribe
def transcribe(audio, state=""):
    #time.sleep(1) #improves quality of transcription
    """ Speech to text function using the pipeline that we defined """
    text = p(audio)["text"]
    state += text + " "
    return state, state
```

```python
# we need input, output and interface components for gradio
gr.Interface(
    fn=transcribe,
    inputs=[
        gr.components.Audio(source="microphone", type="filepath", streaming=True),
        "state"
    ],
    outputs=[
        "textbox",
        "state"
    ],
    live=True).launch()
```
